# Surface temperature from thermal sensors

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

```

### Sample 1

```python
plt.rcParams.update({'font.size': 15})
plt.gca().yaxis.set_major_formatter(FormatStrFormatter('%.1f'))


# Time in minutes
time = [0.00, 1.00, 2.00, 3.00, 4.00, 5.00, 6.00, 7.00, 8.00, 9.00, 10.00]

# Temperatures in Celsius at different angles
temp_90 =  [13.6, 13.6, 13.6, 13.4, 13.3, 13.5, 13.4, 13.4, 13.4, 13.4, 13.4]
temp_0 = [14.4, 14.3, 14.3, 14.4, 14.3, 14.3, 14.4, 14.3, 14.3, 14.2, 14.3]
temp_270 = [14.4, 14.4, 14.3, 14.4, 14.3, 14.3, 14.3, 14.4, 14.3, 14.3, 14.3]
temp_180 = [13.6, 13.6, 13.6, 13.3, 13.6, 13.5, 13.6, 13.4, 13.5, 13.3, 13.5]

# Plotting
plt.plot(time, temp_0, label= r"$0^\circ$", marker='o')
plt.plot(time, temp_90, label=r"$90^\circ$ (top)", marker='o')
plt.plot(time, temp_180, label=r"$180^\circ$", marker='o')
plt.plot(time, temp_270, label=r"$270^\circ (bottom)$", marker='o')
plt.legend()
plt.xlabel("Time (min)")
plt.ylabel(r"Temperature ($^\circ$C)")
plt.grid(True)

```
